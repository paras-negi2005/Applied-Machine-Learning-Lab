**Write a program to create an array [10, 5, 7, 20, 15, -9, 33], print it, Find the maximum and minimum Value of the array**

```python
import numpy as np

# Create array
arr = np.array([10, 5, 7, 20, 15, -9, 33])

# Print array
print("Array:", arr)

# Find maximum and minimum
print("Maximum value:", arr.max())
print("Minimum value:", arr.min())
```

**Write a program to accept an array of elements and find the maximum without using max function**

```python
# Accept array elements from user
arr = list(map(int, input("Enter elements separated by space: ").split()))

# Assume first element is maximum
maximum = arr[0]

# Find maximum manually
for num in arr:
    if num > maximum:
        maximum = num

print("Maximum value:", maximum)
```

**Write a program to accept two matrices [[1,2],[3,4]],   [[2,4],[6,8]]and find the sum and subtraction of two matrices**

```python
# Define matrices
A = np.array([[1, 2],[3, 4]])

B = np.array([[2, 4],[6, 8]])

# Sum of matrices
sum_matrix = A + B

# Subtraction of matrices
sub_matrix = A - B

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nSum of matrices:\n", sum_matrix)
print("\nSubtraction of matrices:\n", sub_matrix)
```

**Write a program to generate a ranodom matrix of size 5 X 6. Add another row containing random elements to it. Find the sum of each row.**

```python
# Generate random 5x6 matrix
matrix = np.random.randint(1, 50, size=(5, 6))

print("Original 5x6 Matrix:\n", matrix)

# Generate another random row
new_row = np.random.randint(1, 50, size=(1, 6))

print("\nNew Random Row:\n", new_row)

# Add row to matrix
matrix = np.vstack((matrix, new_row))

print("\nMatrix after adding new row:\n", matrix)

# Sum of each row
row_sum = matrix.sum(axis=1)

print("\nSum of each row:\n", row_sum)
```

**Write a program to generate a ranodm integer matrix of size 5 X 6. Add another column containing random integers to it. Compute the sine of each element and print it**

```python
# Generate random 5x6 integer matrix
matrix = np.random.randint(1, 50, size=(5, 6))

print("Original 5x6 Matrix:\n", matrix)

# Generate random column
new_column = np.random.randint(1, 50, size=(5, 1))

print("\nNew Random Column:\n", new_column)

# Add column to matrix
matrix = np.hstack((matrix, new_column))

print("\nMatrix after adding new column:\n", matrix)

# Compute sine of each element
sine_matrix = np.sin(matrix)

print("\nSine of each element:\n", sine_matrix)
```

**Write a program to generate a ranodm matrix of size 5 X 6. Consider the 3rd and 4th row; and 2nd, 3rd and 4th column and print the resultant matrix using slicing**

```python
# Generate random 5x6 matrix
matrix = np.random.randint(1, 50, size=(5, 6))

print("Original Matrix:\n", matrix)

# Select 3rd & 4th rows and 2nd, 3rd & 4th columns
# (Indexing starts from 0)
result = matrix[2:4, 1:4]

print("\nResultant Matrix after slicing:\n", result)
```
